```python
import pandas as pd
df = pd.read_csv(r"C:\\Users\\ASUS\\Downloads\\archive\\Iris.csv")
species_list = ["Iris-setosa", "Iris-virginica", "Iris-versicolor"]
modified_dfs = {}
for species in species_list:
    df_copy = df.copy()
    modified_species = []
    for sampledata in df_copy['Species']:
        if sampledata == species:
            modified_species.append(1)
        else:
            modified_species.append(0)
    df_copy['Species'] = modified_species
    modified_dfs[species] = df_copy
    
for species, modified_df in modified_dfs.items():
    print(f"\nTable for species: {species}")
    print(modified_df)

```

# Data Exploration:
Calculate basic summary statistics for the numeric columns (mean, median, min, max, standard deviation).


```python
import pandas as pd
df = pd.read_csv(r"C:\\Users\\ASUS\\Downloads\\archive\\Iris.csv")

summary_stats = {
    'Mean': df.mean(),
    'Median': df.median(),
    'Min': df.min(),
    'Max': df.max(),
    'Standard Deviation': df.std()
}
summary_df = pd.DataFrame(summary_stats)


print(summary_df)
```

```python
import pandas as pd
df = pd.read_csv(r"C:\\Users\\ASUS\\Downloads\\archive\\Iris.csv")

numeric_df = df.select_dtypes(include='number')

summary_stats = {
   
    'Min': numeric_df.min(),
    'Max': numeric_df.max(),
    'Standard Deviation': numeric_df.std()
}

summary_df = pd.DataFrame(summary_stats)

print(summary_df)
```
